```python
import os
import shutil
import pandas as pd

# Set the directory you want to start from
rootDir = 'C:\\Users\\nriveira\\bird_videos\\2023_03_31'

# Set the directory you want to move the files to
moveDir = 'C:\\Users\\nriveira\\bird_videos\\concat'
```

```python
# Cycle through a given directory and subdirectory and rename all videos and store in one place
for dirName, subdirList, fileList in os.walk(rootDir):
    #print('Found directory: %s' % dirName)
    dirSplit = dirName.split('\\')
    year = [idx for idx,match in enumerate(dirSplit) if match.startswith('2023')][0]

    for fname in fileList:
        if fname.endswith('.avi') and not fname.startswith('.') and not dirSplit[year+2].startswith('My_WebCam'):
            vid_title = dirSplit[year] + '_' + dirSplit[year+1] + '_' + dirSplit[year+2] + '_' + fname
            print('\t%s move to %s' % (os.path.join(dirName, fname), os.path.join(moveDir, vid_title)))
            shutil.copy2(os.path.join(dirName, fname), os.path.join(moveDir, vid_title))
```
